# 2. 알고리즘 설명
주어진 코드는 퀵 정렬(Quick Sort) 알고리즘의 일부분을 구현한 것이다. 퀵 정렬은 분할 정복(divide and conquer) 방법을 사용하여 리스트를 정렬하는 효율적인 알고리즘 중 하나다. 주어진 리스트를 특정한 기준값(pivot)을 기준으로 두 개의 부분 리스트로 분할하고, 각 부분 리스트를 재귀적으로 정렬하여 최종적으로 정렬된 리스트를 얻는 방식으로 동작한다.

# 4. 코드 개요
분할(Divide):주어진 리스트를 특정한 기준값(pivot)을 기준으로 두 개의 부분 리스트로 분할한다. 이 때, pivot보다 작은 값은 pivot의 왼쪽에 위치하고, 큰 값은 pivot의 오른쪽에 위치하도록 한다.

정복(Conquer):분할된 각 부분 리스트에 대해 재귀적으로 퀵 정렬을 호출하여 정렬한다.

```python
# 5. 알고리즘 코드
def quick_sort(A, left, right):
    if left<right:
        mid = partition(A, left, right)
        quick_sort(A, left, mid-1)
        quick_sort(A, mid+1, right)
```

```python
# 6. 테스트 코드
def quick_sort(A, left, right):
    if left < right:
        mid = partition(A, left, right)
        quick_sort(A, left, mid - 1)
        quick_sort(A, mid + 1, right)

def partition(A, left, right):
    pivot = A[(left + right) // 2]
    i = left - 1
    j = right + 1
    while True:
        i += 1
        while A[i] < pivot:
            i += 1

        j -= 1
        while A[j] > pivot:
            j -= 1

        if i >= j:
            return j

        A[i], A[j] = A[j], A[i]

test_list = [38, 27, 43, 3, 9, 82, 10]

print("Original: ", test_list)

quick_sort(test_list, 0, len(test_list) - 1)

print("QuickSort:", test_list)
```
