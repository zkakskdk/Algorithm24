#### 손으로 푼 예제


#### 코드 개요
1. 리스트 A입력
2. 리스트 A의 크기를 변수에 저장
3. A의 1번 주소의 원소부터 key에 저장
4. 반복문을 이용하여 정렬

```python
def insetion_sort(A) :
    n = len(A)
    for i in range(1, n) :
        key = A[i]
        j = i-1
        while j >= 0 and A[j] > key :
            A[j+1] = A[j]
            j = j-1
        A[j+1] = key
        printStep(A, j)    
```

```python
def insetion_sort(A) :
    n = len(A)
    for i in range(1, n) :
        key = A[i]
        j = i-1
        while j >= 0 and A[j] > key :
            A[j+1] = A[j]
            j = j-1
        A[j+1] = key
        printStep(A, j) 

def printStep(A, j) :
    for i in range(len(A)) :
        print(A[i], end = "  "  )


def main () :
    temp = input("A 리스트를 공백 문자를 기준으로 띄어서 입력 >>").split()
    temp = list(map(int, temp))
    print(insetion_sort(temp))

    return 0

if __name__ == "__main__":
    main()
```
