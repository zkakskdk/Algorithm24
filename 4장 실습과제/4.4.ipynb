## 2) 알고리즘 설명
1. 진입 차수 초기화 : 먼저, 모든 노드의 진입 차수를 0으로 초기화한다. 진입 차수란 특정 노드로 들어오는 간선의 수를 의미한다. 이를 위해 딕셔너리 inDeg를 사용하며, 그래프의 모든 노드에 대해 0을 할당한다.
2. 진입 차수 계산 : 그래프를 순회하면서 각 노드에 연결된 다른 노드들의 진입 차수를 계산한다. 이를 위해 모든 노드 v에 대해 graph[v]를 순회하며 해당 노드로 들어오는 간선의 수를 증가시킨다.
3. 진입 차수가 0인 노드 찾기 : 위상 정렬을 시작하기 위해 진입 차수가 0인 모든 노드를 찾아 vlist에 추가한다. 진입 차수가 0이라는 것은 해당 노드로 들어오는 간선이 없다는 의미다.
4. 위상 정렬 수행 : vlist에서 노드를 하나씩 꺼내어 출력한다. 그리고 해당 노드와 연결된 모든 노드들의 진입 차수를 감소시킨다.
진입 차수가 0이 된 노드는 vlist에 추가되며, 이를 통해 위상 정렬을 수행할 수 있다. 이 과정을 반복하여 vlist가 빌 때까지 진행한다.

## 4) 코드 개요
1. 진입 차수 초기화 : 먼저, 그래프의 모든 노드에 대한 진입 차수를 0으로 초기화한다. 이를 위해 inDeg라는 딕셔너리를 사용한다.
2. 진입 차수 계산 : 이후, 그래프의 각 노드에 대해 연결된 다른 노드들의 진입 차수를 계산한다. 이를 위해 모든 노드 v에 대해 graph[v]를 순회하면서 각 연결된 노드 u의 진입 차수를 증가시킨다.
3. 진입 차수가 0인 노드 찾기 : 진입 차수가 0인 노드들을 찾아 vlist에 추가한다. 이들은 위상 정렬을 시작할 수 있는 노드다.
4. 위상 정렬 수행 : vlist에서 노드를 꺼내어 출력한다. 그리고 해당 노드와 연결된 모든 노드들의 진입 차수를 감소시킨다.
감소된 진입 차수가 0이 되는 경우, 해당 노드를 vlist에 추가한다. 이 과정을 반복하여 모든 노드가 처리될 때까지 진행한다.

```python
## 5. 알고리즘 코드
def topological_sort(graph) :
    inDeg = {}
    for v in graph :
        inDeg[v] = 0
    for v in graph :
        for u in graph[v]:
            inDeg[u] += 1

    vlist = []
    for v in graph :
        if inDeg[v]==0 :
            vlist.append(v)

    while vlist :
        v = vlist.pop()
        print(v, end=' ')

    for u in graph[v] :
        inDeg[u] -= 1
        if inDeg[u]==0 :
            vlist.append(u)
        
```

```python
## 6. 테스트 코드
def topological_sort(graph):
    inDeg = {}
    for v in graph:
        inDeg[v] = 0
    for v in graph:
        for u in graph[v]:
            inDeg[u] += 1
    
    vlist = []
    for v in graph:
        if inDeg[v] == 0:
            vlist.append(v)
    
    result = []
    
    while vlist:
        v = vlist.pop()
        result.append(v)
        
        for u in graph[v]:
            inDeg[u] -= 1
            if inDeg[u] == 0:
                vlist.append(u)
    
    return result

test_graph = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['D'],
    'D': ['E'],
    'E': []
}

sorted_nodes = topological_sort(test_graph)

print("Topological order:", sorted_nodes)
```
