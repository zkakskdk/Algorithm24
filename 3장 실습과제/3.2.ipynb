## 1)문제 정의
정렬되지 않은 리스트의 탐색 문제
리스트 n개의 항목이 들어 있다. 이 리스트에 "탐색키"를 가진 항목을 찾아라.
만약 찾는 항목이 있으면 그 항목의 인덱스를 반환하고, 없으면 -1을 반환하라.
단, 리스트의 항목들은 정렬되어 있지 않다.

## 2)알고리즘 설명
같이 항목을 하나씩 검사하다가 탐색에 성공하면 해당 항목의 인덱스(3)를 반환한다. 찾는 항목이 없으면 -1을 반환한다. 순차 탐색은 특히 정렬되지 않은 리스트에서도 원하는 항목을 찾을 수 있는 방법이다.

## 4)알고리즘 코드개요
1. 리스트 A의 길이만큼 반복하는 for 루프를 실행한다.
2. 각 반복에서 현재 요소 A[i]와 탐색키 key를 비교한다.
3. 현재 요소와 탐색키가 일치하면 현재 인덱스 i를 반환하고 알고리즘을 종료한다.
4. 리스트를 모두 탐색한 후에도 일치하는 요소를 찾지 못하면 -1을 반환한다.

```python
## 5)알고리즘 코드
def sequenttial_search(A, key):
    for i in range(len(A)) :
        if A[i] == key :
            return i
    return -1
```

```python
## 6) 테스트 코드
def sequential_search(A, key):
    for i in range(len(A)):
        if A[i] == key:
            return i
    return -1

# 사용자로부터 리스트 입력 받기
A = list(map(int, input("리스트를 입력하세요 (공백으로 구분하여): ").split()))

# 사용자로부터 탐색키 입력 받기
key = int(input("탐색할 키를 입력하세요: "))
print(A)
print(key)
# 함수 호출하여 결과 출력
result = sequential_search(A, key)

if result != -1:
    print(f"탐색키 {key}는 리스트에서 인덱스 {result}에 있습니다.", key, result)
else:
    print(f"탐색키 {key}를 찾을 수 없습니다.", key)
```

## 8) 복잡도 분석
이 알고리즘의 시간 복잡도는 O(n). 여기서 n은 리스트 A의 길이를 나타낸다.
선형 탐색은 리스트를 처음부터 끝까지 하나씩 순회하면서 탐색키를 찾는 방법이기 때문에, 최악의 경우에는 리스트의 모든 항목을 한 번씩 확인해야 합니다. 따라서 리스트의 길이에 비례하는 선형 시간이 걸리게 된다.
이러한 이유로 선형 탐색은 대부분의 경우에는 비교적 간단하고 직관적이지만, 매우 큰 리스트에서는 비효율적일 수 있다.
